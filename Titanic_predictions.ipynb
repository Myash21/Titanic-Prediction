```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
```

```python
pipeline1 = Pipeline([("Scaling",StandardScaler()), ("logistic",LogisticRegression())])
```

```python
param_grid = {
    'logistic__penalty': ['l1', 'l2'],  
    'logistic__C': [0.1, 1.0, 10.0],  
    'logistic__solver': ['liblinear', 'lbfgs'],
    'logistic__max_iter': [100, 500, 1000] 
}
```

```python
grid_search = GridSearchCV(pipeline1, param_grid, cv = 5, scoring = "accuracy")
```
